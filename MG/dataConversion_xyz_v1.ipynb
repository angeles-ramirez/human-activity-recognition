```python
 # Import Dependencies
import pandas as pd
import numpy as np
```

```python
output_df = []

Prof_text = ["MG", "GS", "RS","AG", "JD"]
Act_text = ["stand","sit", "laying", "walk"]
for prof in Prof_text:
    for act in Act_text:
        csv_path = "_"+act+"_"+prof+"_acc.csv"
        
        z_file = "z"+csv_path  
        y_file = "y"+csv_path
        x_file = "x"+csv_path
        
        try:
            z_df = pd.read_csv(z_file,header=None)
            z_df.columns = ["Date_time", "Z"]
            y_df = pd.read_csv(y_file, header=None)
            y_df.columns = ["Date_time", "Y"]
            y_df
            x_df = pd.read_csv(x_file, header=None)
            x_df.columns = ["Date_time", "X"]
            xy_df = x_df.merge(y_df, on="Date_time")
            xyz_df = xy_df.merge(z_df, on="Date_time")
            xyz_df["Date_time"]= pd.to_datetime(xyz_df["Date_time"])
            xyz_df['ts'] = xyz_df[['Date_time']].apply(lambda x: x[0].timestamp(), axis=1).astype(int)
            xyz_df["Profile"] = prof
            xyz_df["Activity"] = act
            
            out_path = act+"_"+prof+"_acc.csv"
            xyz_df.to_csv(out_path)
           
            
        except FileNotFoundError:
             print(csv_path+" file format is missing")
        

xyz_df
```

```python
Prof_text = ["MG", "GS", "RS","AG", "JD"]
Act_text = ["stand","sit", "laying", "walk"]
dfs = []
for prof in Prof_text:
    for act in Act_text:
        try:
            csv_path = act+"_"+prof+"_acc.csv"
            dfs.append(pd.read_csv(csv_path))
        except FileNotFoundError:
             print(csv_path+" file not found")    
            
big_frame = pd.concat(dfs, ignore_index=True)    
big_frame
```
